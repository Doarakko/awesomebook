## 2-2 条件指定による抽出

```python
reserve_tb[(reserve_tb.checkin_date >= '2016-10-12') & (reserve_tb.checkin_date <= '2016-10-13')]
```

### Bad
- 可読性が低い
- テーブル名や変数名が変更された場合の変更箇所が多い

```python
reserve_tb.query(" '2016-10-12' <= checkin_date <= '2016-10-13' ")
```

### Good
- 1行で書ける
- SQL に慣れている人にも読みやすい
